```python
# =====================================
# 1. User Login Check
# =====================================
# Hardcoding credentials for now (not ideal for production)

stored_user = "admin"
stored_pass = "1234"

# Simulating user input (in real case, this would come from input())
entered_user = "admin"
entered_pass = "1234"

# Basic comparison
if entered_user == stored_user:
    if entered_pass == stored_pass:
        print("Login Successful")
    else:
        print("Invalid Credentials")  # password mismatch
else:
    print("Invalid Credentials")  # username mismatch

# NOTE:
# Conditions could be combined in one line,
# but this version is easier to read and debug.


# =====================================
# 2. Pass / Fail Analyzer
# =====================================

student_marks = [45, 78, 90, 33, 60]  # sample dataset

passed = 0
failed = 0

# Looping manually for clarity
for m in student_marks:
    if m >= 50:
        passed += 1
    else:
        failed += 1

print("\nPass / Fail Analysis")
print("Passed Students:", passed)
print("Failed Students:", failed)

# Can later add pass percentage if needed


# =====================================
# 3. Simple Data Cleaner
# =====================================

raw_names = [" Alice ", "bob", " CHARLIE "]
processed_names = []

for n in raw_names:
    cleaned = n.strip()      # remove extra spaces
    cleaned = cleaned.lower()  # convert to lowercase
    processed_names.append(cleaned)

print("\nCleaned Names List:")
print(processed_names)

# Step-by-step cleaning makes debugging easier


# =====================================
# 4. Message Length Analyzer
# =====================================

msgs = ["Hi", "Welcome to the platform", "OK"]

print("\nMessage Length Analysis:")

for msg in msgs:
    msg_length = len(msg)
    print("Message:", "'" + msg + "'", "| Length:", msg_length)

    # Check if message is too long
    if msg_length > 10:
        print("⚠️ This message is longer than 10 characters")
    else:
        pass  # no warning needed


# =====================================
# 5. Error Message Detector
# =====================================

log_entries = ["INFO", "ERROR", "WARNING", "ERROR"]

# Using count() for quick check
total_errors = log_entries.count("ERROR")

print("\nError Log Analysis:")
print("Total ERROR entries:", total_errors)

```
